<h3>Sorting & Describing Data</h3>

```python
# Simple sorting
df.sort_values('Species', ascending=True)
df.sort_values(['Species','PetalLengthCm'], ascending=False)
```

<h3>Changing Data</h3>

```python
# Adding total length and total width columns
df['TotalLengthCm'] = df['SepalLengthCm'] + df['PetalLengthCm']
df['TotalWidthCm'] = df['SepalWidthCm'] + df['PetalWidthCm']

df.head(5)
```

```python
# Rearranging the columns
df = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm',
         'PetalWidthCm','TotalLengthCm','TotalWidthCm','Species']]

df.head(1)
```

```python
# Dropping columns
df = df.drop(columns=['TotalLengthCm','TotalWidthCm'], axis=1)

df.head(5)
```

<h3>Visualize Data</h3>

```python
# Plotting data using seaborn & matplotlib
import seaborn as sns 
import matplotlib.pyplot as plt
  
sns.set_style("darkgrid") 

sns.FacetGrid(df, hue ="Species",  
              height = 6).map(plt.scatter,  
                              'SepalLengthCm',  
                              'PetalLengthCm').add_legend() 
```

<h1>Scikit-Learn Practice: Machine Learning Library</h1>

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting df between X and Y
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

<h3>Tuning Hyper Parameters</h3>

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Initiating classifier
knn = KNeighborsClassifier()
```

```python
# Setting hyperparameters
krange = list(range(1,5))
weights_opt = ['uniform','distance']

param_grid = {'n_neighbors': krange,
              'weights': weights_opt}
```

```python
# Initialize grid search
grid = GridSearchCV(knn, param_grid, cv=6, scoring='accuracy')
grid
```

<h3>Fitting KNN Model</h3>

```python
# Fitting sets into grid to search for best estimator of KNN 
grid.fit(X_train, y_train)
```

```python
# Best hyper parameters based on grid search
grid.best_params_
```

```python
# Using model for predictions on test data
pred = grid.predict(X_test)
```

```python
pred
```

<h3>Model Accuracy</h3>

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```
